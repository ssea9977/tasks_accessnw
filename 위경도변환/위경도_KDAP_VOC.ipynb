```python
# 필요 라이브러리
import pyproj
import numpy as np
import pandas as pd
```

```python
# 데이터 가져오기
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 50)

raw_data = pd.read_excel('Daily_5G_LTE_VOC.xlsx', header=2)
raw_data.head()
```

```python
df = raw_data[['위도값(주사용기지국01(음성))', '경도값(주사용기지국01(음성))', 'UTMK_X', 'UTMK_Y']]
df.head()
```

```python
# 도분초 위경도를 WSG84 위경도 형식으로 변환하는 함수 
def transform(lat_list, long_list):
    lats = []
    longs = []
    for i in range(len(lat_list)):
        if lat_list[i][0] != '_':
            lat_do = int(lat_list[i][0])
            lat_bun = int(lat_list[i][1])
            lat_cho = float(lat_list[i][2])
            lat = round(lat_do + ((lat_bun/60) + (lat_cho/3600)), 6)
            lats.append(lat)
        else:
            lats.append(lat_list[i])

    for i in range(len(long_list)):
        if long_list[i][0] != '_':
            long_do = int(long_list[i][0])
            long_bun = int(long_list[i][1])
            long_cho = float(long_list[i][2])
            long = round(long_do + ((long_bun/60) + (long_cho/3600)), 6)
            longs.append(long)
        else:
            longs.append(long_list[i])
    return lats, longs
```

```python
# UTMK 위경도를 WSG84 위경도 형식으로 변환하는 함수 
def project_array(coord, p1_type, p2_type):
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])
    return np.dstack([fx, fy])[0]
```

```python
# 함수에 적용할 Input 데이터 정리
lat_list = df['위도값(주사용기지국01(음성))'].str.split("-")
long_list = df['경도값(주사용기지국01(음성))'].str.split("-")

df['UTMK_X'] = pd.to_numeric(df['UTMK_X'], errors="coerce")
df['UTMK_Y'] = pd.to_numeric(df['UTMK_Y'], errors="coerce")

coord = np.array(df.iloc[:,2:])

p1_type = "epsg:5178"
p2_type = "epsg:4326"

# transform(), project_array() 함수 실행
result = transform(lat_list, long_list)
result2 = project_array(coord, p1_type, p2_type)

# WSG84 위경도 결과값 컬럼 생성 
df['주사용기지국_위도'] = result[0]
df['주사용기지국_경도'] = result[1]
df['UTMK_경도'] = result2[:, 1]
df['UTMK_위도'] = result2[:, 0]
```
